```python
from scipy.io import wavfile
import numpy as np
import pylab as pl
from sklearn.decomposition import FastICA
```

```python
fs_1, voice_1 = wavfile.read("G:\\speech segregation\\Blind-Source-Seperation-master\\Blind-Source-Seperation-master\\mix_type_1_1.wav")
fs_2, voice_2 = wavfile.read("G:\\speech segregation\\Blind-Source-Seperation-master\\Blind-Source-Seperation-master\\mix_type_1_2.wav")
m, = voice_1.shape
voice_2 = voice_2[:m]
```

```python

# and IPython.display for audio output
import IPython.display
# Librosa for audio
import librosa
# And the display module for visualization
import librosa.display
```

```python
# Play it back!
IPython.display.Audio(data=voice_1, rate=fs_1)
```

```python
# Play it back!
IPython.display.Audio(data=voice_2, rate=fs_2)
```

```python
S = [voice_1, voice_2]
IPython.display.Audio(data=S, rate=fs_2)
```

```python
S = np.c_[voice_1, voice_2]
A = np.array([[1, 1], [0.5, 2]])  # Mixing matrix
X = np.dot(S, A.T)  # Generate observations
```

```python
# Compute ICA
ica = FastICA()
S_ = ica.fit(X).transform(X)  # Get the estimated sources
A_ = ica.mixing_  # Get estimated mixing matrix
np.allclose(X, np.dot(S_, A_.T))
```

```python
multiply_factor = 1000000 ;

temp_output_1 = multiply_factor*S_[:,0]
temp_output_2 = multiply_factor*S_[:,1]

wavfile.write("G:\\speech segregation\\Blind-Source-Seperation-master\\Blind-Source-Seperation-master\\Seperated_1new" + ".wav", fs_2, temp_output_1.astype(np.int16))
wavfile.write("G:\\speech segregation\\Blind-Source-Seperation-master\\Blind-Source-Seperation-master\\Seperated_2new" + ".wav", fs_2, temp_output_2.astype(np.int16))
```
